```python
zip_file = [f for f in os.listdir("artifacts/data_ingestion") if f.endswith(".zip")][0]
print(f"zip_file content: {zip_file}")
unzip_path = "./notebooks/artifacts/data_ingestion/data"
os.makedirs(unzip_path, exist_ok=True)
with zipfile.ZipFile(os.path.join("artifacts/data_ingestion", zip_file), 'r') as zip_ref:
    for item in zip_ref.namelist():
        print(f"item content: {item}")
        zip_ref.extract(item, unzip_path)
        
    print(f"Extracted {item} in {unzip_path}")

print(f"Extracted {zip_file} into {unzip_path}")    
```
